Question 3 : Create a Bump Chart to show ranking of Departments based on Treatment Cost over months.


  -To create **Question 3: Bump Chart – Ranking of Departments based on Treatment Cost over Months** in **Tableau**, follow these steps:

### Step 1: Connect the Dataset

* Open Tableau → Connect to **Healthcare_Advanced_Tableau_Assignment_Dataset_v2.csv**

### Step 2: Create Month Field

* Drag **AdmissionDate** → Columns
* Click the dropdown → Select **Month (Discrete)**

### Step 3: Add Treatment Cost

* Drag **TreatmentCost** → Rows
* Change aggregation to **SUM(TreatmentCost)**

### Step 4: Add Department

* Drag **Department** → Marks → **Color**
* Also drag **Department** → **Detail**

### Step 5: Create Rank Calculation

Go to **Analysis → Create Calculated Field**

Name:
`Department Rank`

Formula:

```tableau
RANK(SUM([TreatmentCost]), 'desc')
```

Click **OK**

### Step 6: Build the Bump Chart

* Remove **SUM(TreatmentCost)** from Rows
* Drag **Department Rank** → Rows
* Right-click Rank axis → **Reverse Axis**
  (Rank 1 appears at the top)

### Step 7: Configure Table Calculation

* Right-click **Department Rank**
  → **Edit Table Calculation**
  → Compute Using → **Department**
  → Restart every → **Month**

### Step 8: Change Marks

* Marks → **Line**
* Drag **Department** → Color
* Drag **SUM(TreatmentCost)** → Label (optional)
* Increase line size for visibility

### Final View

* **X-Axis:** Month (AdmissionDate)
* **Y-Axis:** Department Rank
* **Lines:** Departments
* **Ranking Basis:** Sum of Treatment Cost

This produces a **Bump Chart** showing how department rankings change month by month based on treatment cost.


 Question 4 : Create a Butterfly Chart comparing Male vs Female patient count by Department.


  -To create **Question 4: Butterfly Chart comparing Male vs Female Patient Count by Department** in **TableAU**, follow these steps:

### Step 1: Open Dataset

Connect to:
**Healthcare_Advanced_Tableau_Assignment_Dataset_v2.csv**

---

### Step 2: Create Patient Count

Use:

```tableau
Number of Records
```

(or `COUNT([Patient ID])` if available)

---

### Step 3: Create Male Count Field

Go to **Analysis → Create Calculated Field**

Name:

```tableau
Male Count
```

Formula:

```tableau
IF [Gender] = "Male" THEN 1 END
```

---

### Step 4: Create Female Count Field

Create another calculated field:

Name:

```tableau
Female Count
```

Formula:

```tableau
IF [Gender] = "Female" THEN 1 END
```

---

### Step 5: Create Negative Female Measure

Create:

Name:

```tableau
Female Negative
```

Formula:

```tableau
-COUNT([Female Count])
```

---

### Step 6: Build the Butterfly Chart

* Drag **Department** → Rows
* Drag **COUNT(Male Count)** → Columns
* Drag **Female Negative** → Columns (next to Male Count)
* Right-click second axis → **Dual Axis**
* Synchronize Axis

---

### Step 7: Format

* Marks → **Bar**
* Male → Blue
* Female → Orange/Pink
* Edit Axis → Hide negative signs:

  * Right-click Axis → Format → Numbers → Custom

---

### Final Layout

* **Rows:** Department
* **Left Side:** Female Patient Count
* **Right Side:** Male Patient Count
* **Chart Type:** Butterfly Chart

This will display department-wise comparison of **Male vs Female patient counts** mirrored across the center axis.


 Question 5 :  Create a Dual Axis Chart to compare Average Age and Treatment Cost by Department.

   -To create **Question 5: Dual Axis Chart to compare Average Age and Treatment Cost by Department** in **Tableau**, follow these steps:

### Step 1: Connect the Dataset

Open Tableau and load:
**Healthcare_Advanced_Tableau_Assignment_Dataset_v2.csv**

---

### Step 2: Build the Base Chart

* Drag **Department** → **Columns**

* Right-click **SUM(Age)** → **Measure → Average**

---

### Step 3: Add Treatment Cost

* Drag **Treatment Cost** → Rows
  Place it beside `AVG(Age)`

---

### Step 4: Create Dual Axis

* Right-click **SUM(Treatment Cost)** axis
* Select **Dual Axis**

---

### Step 5: Synchronize Axes (Optional)

* Right-click second axis
* Select **Synchronize Axis**

*(Useful if ranges are comparable; otherwise keep separate scales.)*

---

### Step 6: Format Marks

In the **Marks** card:

For **AVG(Age)**:

* Select **Bar**
* Choose one color

For **SUM(Treatment Cost)**:

* Select **Line**
* Choose a different color

---

### Step 7: Improve Readability

* Show **Label**
* Add **Department** labels
* Rename axes:

  * Left → Average Age
  * Right → Treatment Cost

---

### Final View

* **X-Axis:** Department
* **Primary Axis:** Average Age (Bar)
* **Secondary Axis:** Treatment Cost (Line)
* **Chart Type:** Dual Axis Chart

This visual compares **average patient age** and **total treatment cost** across departments in one chart.


Question 6 : Create a Sunburst Chart showing Department → Diagnosis hierarchy by Treatment Cost

  -To create **Question 6: Sunburst Chart showing Department → Diagnosis hierarchy by Treatment Cost** in **Tableau**, follow these steps:

### Step 1: Connect Dataset

Open Tableau and connect to:
**Healthcare_Advanced_Tableau_Assignment_Dataset_v2.csv**

---

### Step 2: Open a New Worksheet

---

### Step 3: Build the Hierarchy

In the Data Pane:

* Right-click **Department** → **Hierarchy → Create Hierarchy**
* Name it:

```text
Department → Diagnosis
```

* Drag **Diagnosis** into that hierarchy under **Department**

Hierarchy becomes:

```text
Department
   └── Diagnosis
```

---

### Step 4: Create the Sunburst Structure

* Drag **MIN(1)** → Columns
* Drag **MIN(1)** → Rows
* Change **Marks** → **Pie**

---

### Step 5: Add Hierarchy Levels

* Drag **Department** → **Color**
* Drag **Diagnosis** → **Detail**

---

### Step 6: Add Treatment Cost

* Drag **Treatment Cost** → **Angle**
* Drag **Treatment Cost** → **Size**
* Change aggregation to:

```text
SUM(Treatment Cost)
```

---

### Step 7: Create Multiple Rings (Sunburst Effect)

Since Tableau has no native Sunburst chart:

Create calculated field:

**Level**

```tableau id="3fqqpi"
INDEX()
```

Then:

* Duplicate the pie chart using **Dual Axis**
* Use:

  * Inner ring → Department
  * Outer ring → Diagnosis
* Adjust **Size** of each Marks card so rings appear concentric

---

### Step 8: Format

* Show labels:

  * Department
  * Diagnosis
  * SUM(Treatment Cost)
* Hide headers
* Remove gridlines

---

### Final View

* **Inner Ring:** Department
* **Outer Ring:** Diagnosis
* **Angle / Size:** Treatment Cost
* **Chart Type:** Sunburst Chart

This chart shows how **Treatment Cost** is distributed from **Department → Diagnosis** hierarchy.


Question 7 : Create a Dual Axis Chart comparing Treatment Cost and Satisfaction Score by Department.

  -To create **Question 7: Dual Axis Chart comparing Treatment Cost and Satisfaction Score by Department** in **Tableau**, follow these steps:

### Step 1: Connect the Dataset

Open Tableau and load:
**Healthcare_Advanced_Tableau_Assignment_Dataset_v2.csv**

---

### Step 2: Add Department

* Drag **Department** → **Columns**

---

### Step 3: Add Treatment Cost

* Drag **Treatment Cost** → **Rows**
* Keep aggregation as:

```text
SUM(Treatment Cost)
```

Question 8 : Create a Dashboard combining:

Heatmap

Dual Axis Chart

Sunburst Chart

Butterfly Chart


 -To create **Question 8: Dashboard combining Heatmap + Dual Axis Chart + Sunburst Chart + Butterfly Chart** in **Tableau**, follow these steps:

## Step 1: Complete the Required Worksheets

Create these sheets first:

1. **Heatmap**

   * Department × Diagnosis
   * Color → Treatment Cost

2. **Dual Axis Chart**

   * Department
   * Treatment Cost (Bar)
   * Satisfaction Score (Line)

3. **Sunburst Chart**

   * Department → Diagnosis
   * Treatment Cost

4. **Butterfly Chart**

   * Male vs Female Patient Count by Department

Rename sheets clearly:

```text
Heatmap
Dual Axis
Sunburst
Butterfly
```

---

## Step 2: Create Dashboard

* Click **New Dashboard**
* Set **Size → Automatic**
  *(or Desktop Browser: 1300 × 900)*

---

## Step 3: Add Layout Containers

From **Objects**:

* Drag **Vertical Container**
* Inside it add **2 Horizontal Containers**

Layout:

```text
---------------------------------
| Heatmap       | Dual Axis     |
---------------------------------
| Sunburst      | Butterfly     |
---------------------------------
```

---

## Step 4: Place Sheets

Drag worksheets into containers:

Top Row:

* Heatmap
* Dual Axis Chart

Bottom Row:

* Sunburst Chart
* Butterfly Chart

---

## Step 5: Add Interactive Filters

Select each sheet →
**Use as Filter**

This allows:

* Click Department in Heatmap → updates others
* Click Sunburst → filters Dual Axis + Butterfly

---

## Step 6: Add Dashboard Actions

Go to:

```text
Dashboard → Actions
```

Add:

### Filter Action

* Source Sheets → All
* Target Sheets → All
* Run Action → Select

### Highlight Action (Optional)

* Highlight selected department

---

## Step 7: Formatting

* Dashboard Title:

```text
Healthcare Treatment Analytics Dashboard
```

Show:

* Legends
* Filter Panel
* Consistent Fonts
* Hide unnecessary headers

---

## Final Dashboard Structure

| Top Left | Top Right |
| -------- | --------- |
| Heatmap  | Dual Axis |

| Bottom Left | Bottom Right |
| ----------- | ------------ |
| Sunburst    | Butterfly    |

This dashboard combines **cost patterns, satisfaction comparison, diagnosis distribution, and gender analysis** into one interactive healthcare dashboard.
